```python
import numpy as np

# 
R_inner = 1.0  # 
R_outer = 7.5  # 
n_xi = 100     # xi
n_eta = 50     # eta
epsilon=1e-3   # 

# xieta
xi = np.linspace(0, 2*np.pi, n_xi)
eta = np.linspace(0, 1, n_eta)

# xy
x = np.zeros((n_xi, n_eta))
y = np.zeros((n_xi, n_eta))

# 
for i in range(n_xi):
    for j in range(n_eta):
        r = R_inner + (R_outer - R_inner) * (eta[j]**2)
        x[i, j] = r * np.cos(xi[i])
        y[i, j] = r * np.sin(xi[i])

# 
max_iterations = 1000

# 
for iteration in range(max_iterations):
    x_new = x.copy()
    y_new = y.copy()
    
    for i in range(1, n_xi-1):
        for j in range(1, n_eta-1):
            # 
            x_xi = (x[i+1, j] - x[i-1, j]) / 2
            x_eta = (x[i, j+1] - x[i, j-1]) / 2
            y_xi = (y[i+1, j] - y[i-1, j]) / 2
            y_eta = (y[i, j+1] - y[i, j-1]) / 2

            # , , 
            alpha = x_xi**2 + y_xi**2
            beta = x_xi * x_eta + y_xi * y_eta
            gamma = x_eta**2 + y_eta**2

            # 
            x_cross = (x[i+1, j+1] - x[i+1, j-1] - x[i-1, j+1] + x[i-1, j-1]) / 2
            y_cross = (y[i+1, j+1] - y[i+1, j-1] - y[i-1, j+1] + y[i-1, j-1]) / 2

            # 
            b_w = b_e = alpha
            b_s = b_n = gamma
            b_p = b_w + b_e + b_s + b_n
            c_px = -beta * x_cross
            c_py = -beta * y_cross

            # xy
            x_new[i, j] = (b_w*x[i-1, j] + b_e*x[i+1, j] + b_s*x[i, j-1] + b_n*x[i, j+1] + c_px) / b_p
            y_new[i, j] = (b_w*y[i-1, j] + b_e*y[i+1, j] + b_s*y[i, j-1] + b_n*y[i, j+1] + c_py) / b_p

    # 

    # 
    max_change = max(np.max(np.abs(x_new - x)), np.max(np.abs(y_new - y)))
    print(max_change)
    if max_change < epsilon:
        print(f'Convergence achieved after {iteration+1} iterations.')
        break

    # 
    x, y = x_new, y_new

# 
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 8))
for i in range(n_xi):
    plt.plot(x[i, :], y[i, :], 'k')
for j in range(n_eta):
    plt.plot(x[:, j], y[:, j], 'k')
plt.gca().set_aspect('equal', adjustable='box')
plt.title('Updated Grid after Iteration')
plt.show()
```
